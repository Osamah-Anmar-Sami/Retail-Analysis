```python
Season = [] 
for month in clean_df['Month_Name'].values:
    if (month == 'December') or (month == 'January') or (month == 'February'): 
        Season.append('Winter') # Winter Season If Month December, January, February

    elif (month == 'March') or (month == 'April') or (month == 'May'):
        Season.append('Spring') # Spring Season If Month March, April, May

    elif (month == 'June') or (month == 'July') or (month == 'August'):
        Season.append('Summer') # Summer Season If Month June, July, August

    elif (month == 'September') or (month == 'October') or (month == 'November'):
        Season.append('Autumn') # Autum Season If Month Septembert, October, November

clean_df['Season'] = Season # # Create New Column Season 
 
```

```python
age_group = []
for age in clean_df['Age'].values:
    if 20 <= age <= 29:
        age_group.append('Twenties') # Twenties If Age Between 20 And 29
    elif 30 <= age <= 39:
        age_group.append('Thirties') # Thirties If Age Between 30 And 39
    elif 40 <= age <= 49:
        age_group.append('Forties') # Forties If Age Between 40 And 49
    elif 50 <= age <= 59:
        age_group.append('Fifties') # Fifties If Age Between 50 And 59
    elif 60 <= age <= 69:
        age_group.append('Sixties') # Sixties If Age Between 60 And 69
    elif 70 <= age <= 79:
        age_group.append('Seventies')  # Seventies If Age Between 70 And 79
    elif age < 20:
        age_group.append('Youth') # Youth If Age Less Than 20

clean_df['Age_Group'] = age_group # Create New Column Age Group
clean_df
```
